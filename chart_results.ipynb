```python
# params
import os
import sys

IN_BED = os.path.expanduser("~/tmp/data/jump_align/250217/output.bed")
```

```python
# read bed file
bed = []
with open(IN_BED) as f:
    for line in f:
        bed_line = line.strip().split()
        bed.append(bed_line)

"read %d lines" % len(bed)
```

```python
# extract data histograms
import matplotlib.pyplot as plt
import numpy as np

fp = np.array([float(x[7]) for x in bed if x[4] == "FP"])
tp = np.array([float(x[7]) for x in bed if x[4] == "TP"])
```

```python
# markers plot
fig, ax = plt.subplots()
ax.plot(range(0, len(fp)), fp, 'x', markeredgewidth=2, markeredgecolor='red')
plt.title("FP")
plt.show()

fig, ax = plt.subplots()
ax.plot(range(0, len(tp)), tp, 'x', markeredgewidth=2, markeredgecolor='blue')
plt.title("TP")
plt.show()
```

```python
# same, just with a certain conditions
fp_v = 0
tp_v = 0
fp_pc = fp[fp >= fp_v]
tp_pc = tp[tp >= tp_v]


fig, ax = plt.subplots()
ax.plot(range(0, len(fp_pc)), fp_pc, 'x', markeredgewidth=2, markeredgecolor='red')
plt.title("FP >= %d (%f%%)" % (fp_v, (100 * len(fp_pc)/len(fp))))
plt.show()

fig, ax = plt.subplots()
ax.plot(range(0, len(tp_pc)), tp_pc, 'x', markeredgewidth=2, markeredgecolor='blue')
plt.title("TP >= %d (%f%%)" % (tp_v, (100 * len(tp_pc)/len(tp))))
plt.show()
```
